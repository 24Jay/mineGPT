## 1. 使用滑动窗口生成数据

```python
import re
```

```python
text = "LLM learns from next token prediction but exhibit general intelligence"
text = re.split(" ", text)

seq_len = 4

for start in range(len(text)-1):
    for i in range(1, seq_len+1):
        print("context: ", text[start: start+i], "======>", text[start+i])
```
